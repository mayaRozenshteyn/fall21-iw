```python
from facebook_scraper import get_posts
import csv
```

```python
# callback function to enable starting from a specificied location in a 
# university's Facebook timeline in the case where scraping is blocked pior
# to completion
start_url = None
def handle_pagination_url(url):
    global start_url
    start_url = url
    print(url)
```

```python
def scrape_posts(school_name):
  # create csv for a specific university
  file_name = school_name + ".csv"
  # open the file in the write mode
  f = open(file_name, 'w')
  # create the csv writer
  writer = csv.writer(f)

  # iterate through the university's Facebook posts and append posts, post time, 
  # & post comments to the csv
  for post in get_posts(school_name, pages=20, start_url=start_url, request_url_callback=handle_pagination_url, options={"comments": True, "posts_per_page": 20}, cookies="cookies.json"):
      caption_list = []
      if post['text'] is not None and post['text'][:] is not None:
          caption_list.append(post['text'][:].replace("\n", " ").replace(',', ':::'))
          caption_list.append(post['time'])
          comments = post['comments_full'][:]
          for comment in comments:
            caption_list.append(comment['comment_text'].replace("\n", " ").replace(',', ':::'))
          writer.writerow(caption_list)
  
  # close the file
  f.close()
```

```python
# example scraper call to obtain relevant Princeton University Fb posts & comments
scrape_posts("PrincetonU")
```
